# 1. Applying Decision Tree

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import sklearn.tree
```

### Applying Decision Tree to the original data

```python
# Preprocessing
df['Age'].fillna(df['Age'].mean(), inplace = True)
df.replace({'Sex':{'male': 1,'female':0}}, inplace = True)
df['Cabin'] = df.Cabin.fillna('G6')
df.replace({'Survived':{'Yes': 1,'No':0}}, inplace = True)
```

```python
# Features extraction
X = df.drop(["Survived", "Name", "Cabin", "Ticket", "Embarked"], axis = 1)
y = df["Survived"]
```

```python
# Splitting data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20,random_state = 10)
```

```python
# Applying tree algorithm
tree = DecisionTreeClassifier()  
tree.fit(X_train, y_train)   # fitting our model
y_pred = tree.predict(X_test)   # evaluating our model
print("Accuracy : {}".format(round(accuracy_score(y_test, y_pred), 3)))
```

### Applying Decision Tree to preprocessed data

```python
new_df.head()
```

```python
# We need to convert all features to numerical
from sklearn import preprocessing
  
label_encoder = preprocessing.LabelEncoder()
  
new_df['Sex'] = label_encoder.fit_transform(new_df['Sex'])
new_df['Cabin'] = label_encoder.fit_transform(new_df['Cabin'])
new_df['Embarked'] = label_encoder.fit_transform(new_df['Embarked'])
new_df['Title'] = label_encoder.fit_transform(new_df['Title'])
```

```python
new_df.head()
```

```python
# features extraction
X = new_df[["Pclass", "Sex", "Age", "SibSp", "Title", "FamilySize"]]
y = new_df["Survived"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)

tree = DecisionTreeClassifier()
tree.fit(X_train, y_train)
y_pred = tree.predict(X_test)

print("Accuracy : {}".format(round(accuracy_score(y_test, y_pred), 3)))
```

We can see that the Decision Tree Classifier is better and more accurate using the preprocessed data.

# 3. Changing the decision tree parameters

In this case, we are going to change max_leaf_nodes, min_samples_leaf and max_depth parameters

```python
# features extraction
X = new_df[["Pclass", "Sex", "Age", "SibSp", "Title", "FamilySize"]]
y = new_df["Survived"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
tree = DecisionTreeClassifier(max_leaf_nodes = 15, min_samples_leaf = 5, max_depth = 10)
tree.fit(X_train, y_train)
y_pred = tree.predict(X_test)
```

# 4. Calculating the new accuracy of the Decision Tree Classifier

```python
# Calculating the accuracy

print("New accuracy : {}".format(round(accuracy_score(y_test, y_pred), 3)))
```

We can see that the accuracy of our model increases by changing some parameters of the decision tree.

# 5. Random Forest

```python
from sklearn.ensemble import RandomForestClassifier # Importing Random Forest Classifier

# features extraction
X = new_df[["Pclass", "Sex", "Age", "SibSp", "Title", "FamilySize"]]
y = new_df["Survived"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
RFC = RandomForestClassifier() # The number of estimators is 100 by default
RFC.fit(X_train, y_train)
y_pred = RFC.predict(X_test)
```

# 6. Calculating the new accuracy of the Random Forest Classifier

```python
# Calculating the new accuracy

print("Accuracy : {}".format(round(accuracy_score(y_test, y_pred), 3)))
```

The Random Forest Classifier is more accurate than the previous models.

### Changing the number of estimators

```python
RFC = RandomForestClassifier(n_estimators = 150)
RFC.fit(X_train, y_train)
y_pred = RFC.predict(X_test)

print("Accuracy : {}".format(round(accuracy_score(y_test, y_pred), 3)))
```

By increasing the number of estimators, the random forest classifier become more accurate.

### To improve the accuracy of a Random Forest Classifier, we will try to change some of its parameters

```python
RFC = RandomForestClassifier(n_estimators = 100, max_leaf_nodes = 15, min_samples_leaf = 5, max_depth = 10)
RFC.fit(X_train, y_train)
y_pred = RFC.predict(X_test)

print("Accuracy : {}".format(round(accuracy_score(y_test, y_pred), 3)))
```

We can see that the accuracy of our model is better than the previous ones.
